```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

def rec(dx,dy,x,y):
    plt.plot([y-1,y+dy-1,y+dy-1,y-1,y-1],[1-x,1-x,1-x-dx,1-x-dx,1-x])

s = open('out.txt').read().split()

for i in range(int(s[0])):
    if int(s[4*i+1])==0:
        if int(s[4*i+4])==1:
            rec(4,5,int(s[4*i+2]),int(s[4*i+3]))
        else:
            rec(5,4,int(s[4*i+2]),int(s[4*i+3]))
    else:
        rec(int(s[4*i+1]),int(s[4*i+1]),int(s[4*i+2]),int(s[4*i+3]))

rec(26,26,1,1)
# rec(5,4,7,1)


ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(1))
plt.title(f'max {int(s[-1])}')
plt.grid(which='both')
plt.axis('scaled')
plt.show()
```
